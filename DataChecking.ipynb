### Exploring the Dataframe

```python
df1_nodupli=df1.drop_duplicates(subset=["InitialConditions", "seed"])
```

```python
df1_nodupli=df1_nodupli.sort_values(by=['InitialConditions','seed'])
```

```python
df1_nodupli['Halt Type'].unique()
```

```python
df1_nodupli.groupby(by="Halt Type").count()
```

```python
df2_nodupli=df2.drop_duplicates(subset=["InitialConditions", "seed"])
```

```python
df2_nodupli['Halt Type'].unique()
```

```python
df2_nodupli.groupby(by="Halt Type").count()
```

```python
df=pd.concat([df1,df2])
```

```python
df_nodupli=df.drop_duplicates(subset=["InitialConditions", "seed"])
```

```python
df_nodupli['Halt Type'].unique()
```

```python
df_haltresult=df_nodupli.groupby(by="Halt Type")
df_haltresult.count()
```

```python
df_haltresult.mean()
```

```python
df_presentable=df_haltresult.count().iloc[:,[0]]
```

```python
df_presentable=pd.merge(df_presentable,df_haltresult.mean().loc[:,['Total Wealth(Producer)','Total Wealth(Consumer)','Number of Shut Firms',
                                                  'Gini Coefficient(Consumers)','Gini Coefficient(Producers)', 'Total Utility','Wage Rate',
                                                  'Leisure Proportion']],left_index=True, right_index=True)
```

```python
df_presentable.reset_index(inplace=True)
```

```python
df_haltresult.std()
```

```python
df_haltresult.mean()
```

```python
df_presentable
```

```python
print(df_presentable.to_latex(float_format="%.4e",index=False,bold_rows=True,longtable=True,caption="Outcome of the simulation runs",label="tab2"))
```
